```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mouse_vo = pd.read_csv('dados/mouse_vo.csv', usecols=['mouse_vo', 'smiles'])
mouse_vo['log_ld50'] = -np.log(mouse_vo['mouse_vo'])

mouse_vi = pd.read_csv('dados/mouse_vi.csv', usecols=['mouse_vi', 'smiles'])
mouse_vi['log_ld50'] = -np.log(mouse_vi['mouse_vi'])

rat_vo = pd.read_csv('dados/rat_vo.csv', usecols=['rat_vo', 'smiles'])
rat_vo['log_ld50'] = -np.log(rat_vo['rat_vo'])

rat_vi = pd.read_csv('dados/rat_vi.csv', usecols=['rat_vi', 'smiles'])
rat_vi['log_ld50'] = -np.log(rat_vi['rat_vi'])
```

```python
# Configurao da figura 4x1 (apenas uma coluna com 4 linhas)
fig, axes = plt.subplots(2, 2, figsize=(10, 6))  # 4 linhas, 1 coluna

# Funo para plotar distribuies
def plot_distribution(data, ax, title):
    sns.histplot(data['log_ld50'], kde=True, ax=ax)
    ax.set_title(title)
    ax.set_xlim(-15, 10)  # Eixo x de -15 a 5
    #ax.set_ylim(0, 4000)  # Eixo y de 0 a 4000
    ax.set_xlabel('-log10 LD50')
    ax.set_ylabel('Frequncia')

# Linha 1: mouse_vo
plot_distribution(mouse_vo, axes[0,0], 'Camundongo, via oral')

# Linha 2: mouse_vi
plot_distribution(mouse_vi, axes[0,1], 'Camundongo, via intravenosa')

# Linha 3: rat_vo
plot_distribution(rat_vo, axes[1,0], 'Rato, via oral')

# Linha 4: rat_vi
plot_distribution(rat_vi, axes[1,1], 'Rato, via intravenosa')

# Ajustando o layout e exibindo a figura
plt.tight_layout()
plt.show()
```
